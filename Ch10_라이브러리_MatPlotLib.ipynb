# 1. Plot for Line

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
pi = np.pi
x = np.linspace(np.pi * -2, np.pi * 2, 100)
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.grid()
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.grid()
```

```python
plt.plot(x - 0, np.sin(x), color='blue')
plt.plot(x - 1, np.sin(x), color='g') # rgbcmyk
plt.plot(x - 2, np.sin(x), color='0.75') # 0 ~ 1 회색조
plt.plot(x - 3, np.sin(x), color='#FFDD44') # RRGGBB, 00 ~ FF
plt.plot(x - 4, np.sin(x), color=(1.0, 0.2, 0.3)) # RGB TUPLE, 0 ~ 1 사이의 값
plt.grid()
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

plt.plot(x, x + 6, linestyle='-')
plt.plot(x, x + 7, linestyle='--')
plt.plot(x, x + 8, linestyle='-.')
plt.plot(x, x + 9, linestyle=':')

plt.grid()
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.xlim(-2 * np.pi, 2 * np.pi)
plt.ylim(1, -1)

plt.grid()
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.axis([-2 * np.pi, 2 * np.pi, 1, -1])

plt.grid()
```

# 4. Grid Data 처리

```python
x = np.random.randint(0, 50, 10)
y = np.random.randint(50, 100, 30)

print(x)
print(y)
```

```python
# Broadcast를 이용해서, 1차원 Mat을 2 by 3 의 2차원 Mat으로 뻥튀기함
xx, yy = np.meshgrid(x, y)
zz = np.random.randint(0, 5, xx.shape)

xx.shape, yy.shape, zz.shape
```

```python
print(xx)
print(yy)
print(zz)
```

```python
rows = xx.shape[0]
cols = xx.shape[1]

for i in range(rows) :
    for j in range(cols) :
        print("[%s,%s] = %s" %(xx[i][j], yy[i][j], zz[i][j]))
```

```python
np.c_[xx.ravel(), yy.ravel(), zz.ravel()]
```

```python
plt.scatter(xx.ravel(), yy.ravel(), c=zz.ravel(), cmap='RdGy') # Red Gray
plt.colorbar()
```

```python
plt.contour(xx, yy, zz, cmap='RdGy')
plt.colorbar()
```

- 등고선 다운 등고선을 그려보자

```python
import lib.byonda as bnd

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

xx, yy = np.meshgrid(x, y)
zz = bnd.get_contour_values(xx, yy)
```

```python
plt.contour(xx, yy, zz, cmap='RdGy')
plt.colorbar()
```

```python
plt.contourf(xx, yy, zz, cmap='RdGy')
plt.colorbar()
```

- 이 플롯은 색상 단계가 연속적이지 않다.
- 단계를 연속화 시키는 것을 렌더링이라 한다.
- 3차원 데이터를 이미지로 표현할 수 있다. : imshow()

```python
plt.contourf(xx, yy, zz, 100, cmap='RdGy')
plt.colorbar()
```

```python
# origin : imshow는 원점이 아래가 아니다. 이것을 변경해야 한다.
# extent : x, y 좌표를 설정해주어야 한다.
plt.imshow(zz, extent=[0, 5, 0, 5], origin='Lower', cmap='RdGy') 
plt.colorbar()
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(xx, yy, zz, 50, cmap='RdGy')
ax.view_init(50, 35)
```

# 7. Subplots
 - 여러개의 그래프를 동시에 plot

```python
x1 = np.random.randn(100).cumsum()
x2 = np.random.randn(100).cumsum()
x1, x2
```

```python
plt.plot(x1)
plt.plot(x2)
```

```python
plt.subplot(2, 2, 1) # row col idx 
plt.plot(x1)

plt.subplot(2, 2, 4) # row col idx 
plt.plot(x2)
```

```python
plt.subplot(2, 2, 1) # row col idx 
plt.plot(x1, color='r')

plt.subplot(2, 2, 4) # row col idx 
plt.plot(x2, color='b')
```

```python
plt.subplot(2, 2, 1) # row col idx 
plt.plot(x1, color='r', label = 'x1')
plt.legend()

plt.subplot(2, 2, 4) # row col idx 
plt.plot(x2, color='b', label = 'x2')
plt.legend()
```

```python
plt.subplot(2, 2, 1) # row col idx 
plt.plot(x1, color='r', label = 'x1')
plt.legend(fontsize=14)

plt.subplot(2, 2, 4) # row col idx 
plt.plot(x2, color='b', label = 'x2')
plt.legend(fontsize=8)
```

```python
plt.subplot(2, 2, 1) # row col idx 
plt.plot(x1, color='r', label = 'x1')
plt.legend(fontsize=14)

plt.subplot(2, 2, 4) # row col idx 
plt.plot(x2, color='b', linestyle='-.', label = 'x2')
plt.legend(fontsize=8)
```

```python
plt.subplot(2, 2, 1) # row col idx 
plt.plot(x1, color='r', label = 'x1')
plt.legend(fontsize=14)

plt.subplot(2, 2, 2) # row col idx 
plt.plot(x2, color='b', linestyle='-.', label = 'x2')
plt.legend(fontsize=8)

plt.subplot(2, 2, 3) # row col idx 
plt.plot(x2, color='b', linestyle='-.', label = 'x2')
plt.legend(fontsize=8)

plt.subplot(2, 2, 4) # row col idx 
plt.plot(x2, color='b', linestyle='-.', label = 'x2')
plt.legend(fontsize=8)
```
